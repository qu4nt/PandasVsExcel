# Otras estructuras de datos

Recordemos, una estructura de datos es un conjunto de valores (datos) almacenados en un objeto.

En Excel tenemos una tabla que nos permite almacenar la información, es una estructura de datos. En Python, como vimos, existen diversas formas de almacenar datos entre ellas están las listas, tuplas y diccionarios los cuales guardan o compactan la información que puede ser de distintos tipos.

Por ejemplo, en una lista que llamaremos var se puede almacenar números, cadenas de caracteres, tuplas e incluso podemos incluir listas.


```python
var = [1, 2, 'valor3', 4.0, (5, 'valor6')]
```

 Esto implica que una lista puede contener más listas, sería una lista de listas y se parece a una matriz. Por ejemplo,

```python
ll = [[1, 2, 3],
      [4, 5, 6],
      [7, 8, 9]]
```

```python
ll
```

Es algo similar a tener tres filas y tres columnas en Excel.

image.png

¿Cómo accedemos a los elementos de las listas?

Para acceder a un elemento específico de esa lista debemos escribir la ubicación. Por ejemplo, si queremos acceder al elemento en la primera fila y la tercera columna

```python
ll[0][2]
```

¿Por qué 0 y no 1, 2 y no 3?

En Python las posiciones comienzan en 0 entonces si tenemos 3 elementos tendremos las posiciones 0, 1, 2 y así sucesivamente.

Ahora que ya repasamos las listas, debes conocer que en Python también existen otras estructuras de datos que permiten el manejo más eficiente de los datos. 

Comencemos con arrays

## Arrays
Estos objetos son estructuras de datos que representan arreglos de valores, son similares a las listas de listas con la diferencia que los valores deben ser del mismo tipo.

Los arrays son objetos que podemos crear a partir del módulo array de Python y son equivalentes a una matriz de datos numéricos en Excel.


Para crear un arreglo solo debemos importar el módulo array, acceder a la clase array, pasarle el tipo de datos y los valores.

```python
import array
```

```python
array.array( 'd', [1.0, 2.0, 3.14])
```

'd' se refiere a números flotantes

En este caso se tienen distintos tipos de datos y varias formas de almacenamiento en memoria, si no recordamos las letras correspondientes podemos utilizar la ayuda.

```python
help(array.array)
```

En Python también tenemos disponible la biblioteca Numpy que permite crear y realizar operaciones sobre arreglos. Los arreglos o arrays de Numpy son más flexibles porque permiten que sus valores sean de cualquier tipo de objeto en Python.

## Arreglos de Numpy (np.array)



Un array de numpy es un conjunto de valores, que pueden o no ser del mismo tipo, y están indexados por una tupla de números enteros no negativos. El número de dimensiones es el rango del array; su forma es una tupla de enteros que da el tamaño del array a lo largo de cada dimensión.

Podemos inicializar un array de numpy a partir de listas anidadas de Python, y acceder a sus elementos utilizando corchetes.

```python
import numpy as np
```

```python
arreglo = np.array(['Kylie', 'Jenner', 590000000, 'Kanye', 'West', 170000000, 'Roger', 'Federer', 106300000])
```

```python
arreglo[3]
```

Los arreglos de Numpy son eficientes especialmente en los casos en los que se tienen datos numéricos.

📈 Los arrays o arreglos se pueden crear de dos formas, una es con el módulo array y la otra con el método array de Numpy.

Una forma de crear un arreglo es a partir de una lista.

```python
ll_array = np.array(ll)
ll_array
```

## Series de Pandas

Una serie es un arreglo unidimensional, como una columna de Excel. Las series poseen un índice que puede o no indicar la posición de los datos. Trabajar con una Serie es análogo a referenciar una columna de una hoja de Excel.

Para crear una serie debemos importar la biblioteca *pandas* y si por ejemplo, queremos crear una serie a partir de una lista

```python
import pandas as pd

serie_ = pd.Series([2, 4, 6, 8, 10])
```

```python
serie_
```

Las series también las podemos crear a partir de diccionarios, listas o arrays (arreglos).

Ahora, veamos los DataFrames.

## DataFrame

Es la estructura más similar a una hoja de Excel. Se puede generar a partir de diferentes objetos como diccionarios, listas o Series. Arreglos de numpy. Los DataFrames están compuestos por series. 

En aquellos casos en que los datos provengan de fuentes tabulares los DataFrames de *pandas* son esenciales.






```python
import pandas as pd
```

```python
dataframe = pd.DataFrame({'frutas' : ['manzana', 'pera', 'mandarina'], 'precios' : [2, 3, 1]})
```

```python
dataframe
```

```python
type(dataframe)
```

Como se puede observar un DataFrame cuenta con un índice que indica la posición de una observación o de un valor, también cuenta con columnas.
