## **Valor absoluto**

En Excel podemos calcular el valor absoluto a partir de la función ABS.
Por ejemplo, si queremos calcular el valor absoluto de -30



image.png

Ahora, esta misma operación en Python la podemos realizar de dos formas.

1. Con la función `abs`

```python
abs(-30)
```

2. Con el módulo `math`

```python
import math as mt
```

```python
mt.fabs(-30)
```

Unir cadenas de caracteres

**Concatenación de caracteres**

En muchas ocasiones necesitamos unir cadenas para que estén en una sola celda o columna, en Excel lo podemos hacer con la función `concat`.

image.png

En Python esto lo podemos hacer conociendo las propiedades de las cadenas (strings) o mediante `join`. Por ejemplo, 

```python
nombre = 'Luisa'
apellido = 'Martinez'
```

```python
nombre + " " + apellido 
```

```python
sep= " "
sep.join([nombre, apellido])
```

En este caso puedes observar que utilizamos una lista que contiene las dos variables, esto es porque `join` requiere un argumento iterable para el cual utilizaremos un separador y la función join los unirá.



💡 Un iterable es un objeto sobre el que se puede hacer un bucle o iterar con la ayuda de un bucle for. Objetos como listas, tuplas, conjuntos, diccionarios, cadenas, son llamados iterables. Si quieres saber más de esto puedes leer un artículo [aquí](https://www.analyticsvidhya.com/blog/2021/07/everything-you-should-know-about-iterables-and-iterators-in-python-as-a-data-scientist/#:~:text=Iterable%20is%20an%20object%20which,that%20you%20can%20loop%20over).

## **Conteo de valores**

En Excel existe una función que permite contar la cantidad de números en un rango de datos (`CONTAR`), asimismo, existe una función que permite contar los datos dado un criterio (`CONTAR.SI`) y una función que cuenta la cantidad de celdas que no están vacías en un intervalo (`CONTARA`) .

Por ejemplo, si tenemos una columna de valores como los siguientes y utilizamos la función CONTAR, esta nos devuelve el valor 5 que es la cantidad de números.

image.png

Ahora, si queremos saber la cantidad de 6 en la columna utilizamos CONTAR.SI

image.png

O si queremos determinar la cantidad de apariciones de una palabra en específico utilizamos también CONTAR.SI con la palabra entre comillas


image.png

En Python, por ejemplo, si tenemos nuestros datos en una lista podemos utilizar `count` y entre los argumentos el valor que queremos contar. Es decir, count devuelve el número de ocurrencias de un valor dado.

```python
ls = [-30, 6, 9, 3, 6, 'dato', 'fecha']
```

```python
ls.count(6)
```

```python
ls.count('dato')
```

## **Suma de valores**

En Excel para sumar valores podemos utilizar la función `SUMA`, esta función puede tener entre los valores a operar cadenas de caracteres y no los tomará en cuenta. 

image.png

En Python podemos sumar valores utilizando la función `sum`, esta función solo admite valores numéricos. Si probamos con la lista que creamos anteriormente, nos genera un error:

```python
sum(ls)
```

Sabiendo entonces que `sum` admite solo valores numéricos podemos acceder a los valores numéricos en las primeras seis posiciones de la lista y sumarlos.

```python
sum(ls[:5])
```

Nos da el mismo resultado que en Excel. Ahora, la misma operación con la función fsum del módulo math

```python
mt.fsum(ls[:5])
```

Y con la función sum de Numpy

```python
import numpy as np
```

```python
np.sum(ls[:5])
```

**Evaluación de condiciones**

Para evaluar una condición en Excel se utiliza la función SI, se debe plantear la función lógica y los valores en caso de que se cumpla o no la condición.

Por ejemplo, si nuestros datos numéricos corresponden a notas de alguna evaluación y para que este aprobado los valores deben ser mayores a 5, en Excel podemos reflejarlo de la siguiente manera:

image.png

image.png

En Python esto lo podemos replicar a partir de un bucle for que aplicaremos a la parte de la lista que solo contiene números. Este bucle se encarga de recorrer la lista y va aplicando las funciones o evaluando cada valor dependiendo de lo que indiquemos. Para este caso evaluaremos los valores y si es mayor a 5 debemos colocar aprobado, en caso contrario reprobado, entonces

```python
for i in ls[:5]:
  if i > 5:
    print(i, 'Aprobado')
  else:
    print(i, 'Desaprobado')
```

## Otras funciones

### Media 

En Excel podemos calcular la media o promedio a partir de la función PROMEDIO.

image.png

En Python, la media o promedio se puede calcular a partir del módulo statistics, especificamente con la función mean o con la biblioteca Numpy a partir de las funciones average y mean.

```python
import statistics

statistics.mean(ls[:5])
```

```python
np.average(ls[:5])
```

```python
np.mean(ls[:5])
```

### Mínimo

En Excel, el mínimo se puede calcular a partir de la función MIN.

image.png

En Python podemos calcularlo a partir de la función `min` de la biblioteca estándar y con la función `min` de Numpy.

```python
min(ls[:5])
```

```python
np.min(ls[:5])
```

### Máximo

En Excel podemos calcular el máximo a partir de la función MAX.

image.png

Para Python, el máximo se puede calcular con la función max de la biblioteca estándar o de la biblioteca Numpy.

```python
max(ls[:5])
```

```python
np.max(ls[:5])
```
