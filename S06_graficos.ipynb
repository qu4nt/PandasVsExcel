# Gráficos con *pandas*

Los gráficos nos permiten representar nuestra información de forma visual, a apartir de ellos podemos tener un panorama más claro de nuestros valores. En Excel podemos realizar una serie de gráficos como barras, líneas, pie, área y dispersión. 

La biblioteca _pandas_ nos permite realizar gráficos, a continuación veremos cuán sencillo es.


Antes de continuar quiero que sepas que los gráficos dependen del tipo de datos que tengamos por ejemplo, no deberías hacer un gráfico de torta para una variable que almacene el monto de facturas.

Importamos nuestros datos. En este caso utilizaremos nuestros datos sobre los famosos. 

Luego de importarlos debemos cambiar aquella información que nos impide comprender bien nuestros datos.

```python
import pandas as pd
import numpy as np
```

```python
from numpy import datetime64
```

```python
data = pd.read_excel('/content/drive/MyDrive/famosos.xlsx', 
                     dtype={'Nombre': str,
                            'Apellido': str,
                            'Sexo': 'category',
                            'Fecha de nacimiento': datetime64,
                            'Fuente de la riqueza': "category",
                            'Residencia': "category",
                            'Nacionalidad': "category",
                            'Hijos': "int64", 
                            'REdad': "category"} )
```

## Gráficos de líneas

Estos gráficos son buenos para mostrar las tendencias de los datos. Se colocan los valores de la variable en el eje horizontal y se trazan los puntos respectivos a las frecuencias en el eje vertical, luego se unen con líneas. Sirve para mostrar varias variables y se puede elegir entre mostrar la frecuencia o los porcentajes en el eje vertical. 

Los utilizamos para datos numéricos.

```python
data[['edad']].value_counts().sort_index().plot(kind='line')
```

Estos gráficos también son útiles en el caso de que queramos presentar los datos de las series de tiempo.

💡 El gráfico de línea es el gráfico preestablecido en la función plot, por ello si solo escribes

df.variable.plot() se mostrará un gráfico de líneas.

## Gráficos de barras/columna

Este tipo de gráficos consiste en dibujar una barra o un rectángulo para cada una de las modalidades de la variable (o el tiempo en el caso de series cronológicas) dispuestos en el eje horizontal. La barra tendrá una altura igual a su frecuencia o porcentaje (en el eje vertical). 

Este tipo de gráfico es ideal cuando tenemos variables con categorías o existe un número finito de valores para los cuales nos interesa el número de repeticiones en cada caso.

```python
data.Sexo.value_counts().plot(kind='bar');
```

Hay más cantidad de famosos de sexo masculino que de sexo femenino.

## Gráficos circulares

Se trata de un círculo dividido en porciones en forma de cuña, cada una representa un grupo o categoría de la variable. El tamaño de cada trozo es proporcional a la cantidad de valores en cada grupo en comparación con los demás. Estos valores se expresan en términos de porcentaje. Se utilizan para representar variables en las que tenemos categorías o grupos de datos asociados a un(os) determinado(s) valor(es).

```python
data.Sexo.value_counts().plot(kind='pie');
```

El gráfico en este caso representa la misma información que en el ejemplo anterior.

## Gráficos de barras horizontal

Es equivalente al gráfico de barras vertical. En este gráfico se representan las categorías o valores en el eje y, mientras que las frecuencias irán en el eje x.

Es ideal para los casos en los que tenemos categorías o una cantidad finita de valores en nuestros datos.

```python
data.Sexo.value_counts().plot(kind='barh')
```

Aquí observamos que se representa exactamente la misma información que en el caso del diagrama de barras verticales.

## Gráfico de área

Los gráficos de área son gráficos de línea pero con el área debajo de la línea coloreada.

Se utilizan para representar valores numéricos o correspondientes a series de tiempo.

```python
data[['Hijos']].value_counts().sort_index().plot(kind='area')
```

## Histogramas

Es un caso particular del gráfico de barras para variables con valores numéricos. Este gráfico permite ilustrar el comportamiento de los valores agrupados en intervalos de clase, siendo un gráfico de barras compuesto por varios rectángulos adyacentes en el eje horizontal, donde cada intervalo es la base de cada rectángulo; en el eje vertical se marcan las alturas que vienen dadas por las frecuencias (absolutas o relativas). Las barras no llevan espacios entre sí. Un ejemplo muy utilizado de histograma es una pirámide de población.

imagen.png

*Fuente: https://population-pyramid.net/es/pp/espa%C3%B1a*

Permiten visualizar:

* Comportamientos o patrones en los datos.

* Simetría en el conjunto de datos.

* La dispersión entre las clases de la distribución.

* El punto de equilibrio en las frecuencias de las clases.



```python
data[['edad']].plot(kind='hist');
```

Excel también permite hacer histogramas, para estos datos se genera lo siguiente

## Gráfico de dispersión/scatter

Se utiliza cuando se desea mostrar los valores de dos variables cuantitativas para ver en qué medida una variable afecta a la otra, es decir, visualizar la relación entre dos variables, una en el eje x (generalmente la variable independiente) y la otra, en el eje y (la variable dependiente) y luego se grafican los pares de valores con coordenadas (x,y). Así pues, el resultado será un conjunto de puntos "dispersos" alrededor del plano.

```python
data.plot(x='edad', y='Hijos',kind='scatter');
```

## Boxplot

Otro gráfico que puedes hacer con pandas es el boxplot.


Los gráficos de caja y bigotes o boxplot representan visualmente un conjunto de datos numérico a través de divisiones (cuartiles).

Podemos visualizar en este gráfico: rango, datos atípicos, porciones porcentuales en nuestros datos, mediana, mínimo y máximo.

```python
data.edad.plot(kind='box')
```

Este gráfico también lo podemos hacer con 


```python
data.boxplot(column='edad')
```

En este gráfico podemos ver que la edad media está entre 42 o 43 años. También nos muestra que la menor edad está por debajo de 30 y la más alta por encima de 70 años.
